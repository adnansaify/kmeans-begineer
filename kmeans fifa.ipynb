```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("player.csv")
df
```

```python
plt.scatter(df['Age'],df['Overall'])
```

```python
km=KMeans(n_clusters=5)
predicted=km.fit_predict(df[['Age','Overall']])
predicted
```

```python
df['cluster']=predicted
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
df5=df[df.cluster==4]

plt.scatter(df1.Age,df1['Overall'],color='green')
plt.scatter(df2.Age,df2['Overall'],color='black')
plt.scatter(df3.Age,df3['Overall'],color='red')
plt.scatter(df4.Age,df4['Overall'],color='yellow')
plt.scatter(df5.Age,df5['Overall'],color='blue')

plt.xlabel['Age']
plt.ylabel['Overall']
plt.legend
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Overall']])
df['Overall']=scaler.transform(df[['Overall']])

scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
df.head()
```

```python
plt.scatter(df.Age,df['Overall'])
```

```python
see=[]
k_range=range(1,10)
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Overall']])
    see.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('sum of squared error')
plt.plot(k_range,see)
```
